```python
import pickle
from PIL import Image
from matplotlib import pyplot as plt
```

```python
# load the dataframe
df = pickle.load(open('annotation_frame.p','rb'))
```

```python
# display some comlumns
# filename -> filename of the AgNor image
# bbox:
# min_x,min_y-------------
#   |                    |
#   |                    |
#   -------------------max_x,max_y
# label -> number of AgNors visible in in the cell
df.head()
```

```python
print(f'Number of available annotations: {len(df)}')
```

```python
# display classes present
# 11 -> more then 10 AgNORs identifiable

print(sorted(df.label.unique()))
```

```python
# add path to slides here
path_to_image = 'images/'

# show some images
fig, ax = plt.subplots(5,5, figsize = (5,5))
ax = ax.flatten()
for i in range(25):
    row = df.sample()
    # open image
    img = Image.open(path_to_image + row.filename.item()).convert('RGB')
    # crop
    crop = img.crop((row.min_x, row.min_y, row.max_x, row.max_y))
    ax[i].imshow(crop)
    ax[i].axis('off')
    # use the label as a title
    ax[i].set_title(row.label.item())
    # close image again
    img.close()
plt.tight_layout()
```
