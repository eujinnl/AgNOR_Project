```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import pickle
from PIL import Image
import random
from torchvision.transforms import ToTensor
from torchvision.utils import draw_bounding_boxes
import torchvision.transforms.functional as TF






class Dataset(torch.utils.data.Dataset):
    
    def __init__(self, annotations_frame, path_to_slides,  pseudo_epoch_length=1000,crop_size=(128,128), transformations= None):
        super().__init__()
        self.annotations_frame = annotations_frame
        self.path_to_slides = path_to_slides
        self.crop_size = crop_size
        self.pseudo_epoch_length = pseudo_epoch_length
        self.transformations = transformations
        self.df = annotations_frame
        self.images = {}
        self.initialize()
        
    def initialize(self):
        for filename in self.df['filename'].unique():
            img_path = f'{self.path_to_slides}images/{filename}'
            img = Image.open(img_path)
            self.images[filename] = img
            
    def __getitem__(self,index):
        target = {'boxes':[],'labels':[]}
        random_choice =random.choice(self.df['filename'].unique())
        img_w, img_h,  = self.images[random_choice].size
        min_x= random.randint(0,img_w-self.crop_size[0])    #point where crop is made
        min_y= random.randint(0,img_h-self.crop_size[1])    #point where crop is made
        max_x=min_x+self.crop_size[0]
        max_y=min_y+self.crop_size[1]        
        random_patch = self.images[random_choice].crop((min_x,min_y,max_x,max_y))
        x= self.df.loc[(self.df['filename']==random_choice) & (self.df['max_x'] < max_x) & (self.df['min_x'] > min_x)
                       & (self.df['max_y'] < max_y) & (self.df['min_y'] > min_y) & (self.df['label'] != 0)] 
        transform =ToTensor()
        random_patch= transform(random_patch.convert("RGB"))
        random_patch = TF.convert_image_dtype(image=random_patch,dtype=torch.uint8)
        #CHOOSE from annotation frame the correct file and then the cells that are within the parameters
        for i in x.index:
            x.loc[i,"min_x"]-=min_x
            x.loc[i,"max_x"]-=min_x
            x.loc[i,"max_y"]-=min_y
            x.loc[i,"min_y"]-=min_y
            xmin, ymin, xmax, ymax,label = x.loc[i,"min_x"],x.loc[i,"min_y"],x.loc[i,"max_x"],x.loc[i,"max_y"],x.loc[i,'label']
            # bounding box in (xmin, ymin, xmax, ymax) format
            bbox= [xmin,ymin,xmax,ymax]
            target['labels'].append(str(label))
            #bbox = torch.tensor(bbox, dtype=torch.float)
            target['boxes'].append(bbox)
        target['boxes'] = torch.tensor(target['boxes'], dtype=torch.float)
        random_patch=draw_bounding_boxes(random_patch,target['boxes'],labels = target['labels'],
                                             width=1,colors=(255,0,0),fill =True)
        random_patch = TF.convert_image_dtype(image=random_patch,dtype=torch.float32)

            
   
        self.pseudo_epoch_length -= 1
        
        target['labels'] =torch.tensor([int(i) for i in target['labels']],dtype= torch.int64)

        #print(random_patch)
        return random_patch,target
    
      
    def __len__(self):
        if self.pseudo_epoch_length == 0:
            return False
        else:
            return True
        
            
```

```python
# load annotation data and set path to images
path_to_images = ''
anno_data = pickle.load(open('annotation_frame.p','rb'))
#print(anno_data)

# instantiate your dataset, you have to define or import it first
ds = Dataset(annotations_frame=anno_data,
        path_to_slides=path_to_images,
        crop_size=(128,128),
        transformations=None)
```

```python
# load an image and a target
img, target = ds[0]
plt.imshow(img.permute(1, 2, 0))
#print(target['boxes'])
print(target['labels'])
```

# You can use this skript to validate your dataset

```python
# testing function
def test(image: torch.tensor, targets: dict):
    '''
    Function for checking your dataset.

    Input:
    image -> pytorch image tensor
    target -> dict holding the bounding boxes and the labels
    '''
    c = 0
     # check image
    if image.dtype != torch.float32:
        print('Wrong image dtype, needs to be torch.float32')
    elif image.shape[0] != 3 or len(image.shape) != 3:
        print('Wrong image shape, needs to be [3, height, width] for rgb images.')
    else:
        print("Image passed test!")
        c +=1
    # test targets
    # check bounding boxes
    assert 'boxes' in targets, 'the filed "boxes" has to be defined'
    if targets['boxes'].dtype != torch.float32:
        print('Wrong dtype for bounding boxes, has to be torch.float32')
    else:
        if len(targets['boxes']) == 0:
            print('No objects found on the image, try another one')
        elif targets['boxes'].shape[-1] != 4 or len(targets['boxes'].shape) != 2:
            print('Wrong bounding box shape, should be (FloatTensor[N, 4]) (N -> number of instances): the coordinates of the N bounding boxes in [x0, y0, x1, y1] format')
        else:
            print('Bounding boxes passed test!')
            c +=1
    # check labels
    assert 'labels' in targets, 'the filed "labels" has to be defined'
    assert targets['boxes'].shape[0] == targets['labels'].shape[0], 'The first dimensions of the bounding box vector ([N,4]) needs to be the same as the length of the labels vector ([N])'
    
    if targets['labels'].dtype != torch.int64:
        print('Wrong dtype for bounding boxes, has to be torch.float64')
    else:
        if len(targets['labels'].shape) != 1:
            print('Wrong shape for labels, should be [N] (N -> number of instances)')
        elif 0 in targets['labels'] and len(targets['boxes']) != 0:
            print('Illegal label 0 found, 0 is always background. Choose 1 for the cells')
        else:
            print('Labels passed test!')
            c +=1
            
        
    if c == 3:
        print('\nAll checks passed successfully!')
```

```python
# test your input
test(img,target)
```
